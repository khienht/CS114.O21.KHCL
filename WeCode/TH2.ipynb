# dịch rìa

```python
def shift_edge_clockwise(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    top_row = matrix[0]
    bottom_row = matrix[-1]
    left_col = [matrix[i][0] for i in range(1, rows - 1)]
    right_col = [matrix[i][-1] for i in range(1, rows - 1)]

    shifted_matrix = []
    shifted_matrix.append(bottom_row + top_row[1:-1])
    for i in range(1, rows - 1):
        shifted_matrix.append([right_col[i-1]] + matrix[i][1:-1] + [left_col[i-1]])
    shifted_matrix.append(top_row[::-1] + bottom_row[1:-1])

    return shifted_matrix

def shift_edge_counter_clockwise(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    top_row = matrix[0]
    bottom_row = matrix[-1]
    left_col = [matrix[i][0] for i in range(1, rows - 1)]
    right_col = [matrix[i][-1] for i in range(1, rows - 1)]

    shifted_matrix = []
    shifted_matrix.append(top_row[::-1] + bottom_row[1:-1])
    for i in range(1, rows - 1):
        shifted_matrix.append([left_col[i-1]] + matrix[i][1:-1] + [right_col[i-1]])
    shifted_matrix.append(bottom_row + top_row[1:-1])

    return shifted_matrix

def shift_edges(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    new_matrix = matrix
    for i in range(4):  # There are 4 edges
        if i % 2 == 0:
            new_matrix = shift_edge_clockwise(new_matrix)
        else:
            new_matrix = shift_edge_counter_clockwise(new_matrix)
    return new_matrix

# Input
r, c = map(int, input().split())
matrix = [list(map(int, input().split())) for _ in range(r)]

# Output
shifted_matrix = shift_edges(matrix)
for row in shifted_matrix:
    print(' '.join(map(str, row)))
```

# **Chu vi đường biên**

```python
def boundary(m, n, matrix):
    length = 0

    for i in range(m):
        for j in range(n):
            if matrix[i][j] == 1:
                if i == 0 or matrix[i-1][j] == 0:
                    length += 1  # above
                if i == m - 1 or matrix[i+1][j] == 0:
                    length += 1  # underneath
                if j == 0 or matrix[i][j-1] == 0:
                    length += 1  # left
                if j == n - 1 or matrix[i][j+1] == 0:
                    length += 1  # right

    return length

m, n = map(int, input().split())
matrix = []
for _ in range(m):
    row = list(map(int, input().split()))
    matrix.append(row)

print(boundary(m, n, matrix))

```

# **Đếm sao**

```python
def count_stars(m, n, image):
    count = 0

    for i in range(m):
        for j in range(n):
            if image[i][j] == '-':
                count += 1
                mark_star(image, m, n, i, j)

    return count


def mark_star(image, m, n, i, j):
    if i < 0 or i >= m or j < 0 or j >= n or image[i][j] == '#':
      return

    image[i][j] = '#'

    mark_star(image, m, n, i + 1, j)
    mark_star(image, m, n, i - 1, j)
    mark_star(image, m, n, i, j + 1)
    mark_star(image, m, n, i, j - 1)

test_case = 1
while True:
    try:
        m, n = map(int, input().split())
        if m == 0 and n == 0:
            break

        image = [list(input()) for _ in range(m)]

        result = count_stars(m, n, image)
        print(f"Case {test_case}: {result}")

        test_case += 1
    except EOFError:
        break


```

# **Con đò**

```python
l, m = map(int, input().split())
len = l*100
cars = [input() for _ in range(m)]
left = []
right = []

for car in cars:
    length, pos = car.split()
    if pos == "left":
        left.append(int(length))
    else:
        right.append(int(length))

counts=0
current_side = "left"
while left or right:
    if current_side == "left":
        total = 0
        while left and total + left[0] <= len:
            total += left.pop(0)
        counts += 1
        current_side = "right"
    else:
        total = 0
        while right and total + right[0] <= len:
            total += right.pop(0)
        counts += 1
        current_side = "left"

print(counts)
```

# Mảng xém đối xứng

```python
def mang_dx(n, arr):
  mid = (n + 1) // 2
  count=0
  for i in range(mid):
    if arr[i] != arr[n - i - 1] :
      count+=1
      if count>1:
        return False
      if arr[i] == arr[n - i - 2] and arr[i + 1] == arr[n - i - 3]:
        return False
  return True

n = int(input())
arr = [int(input()) for _ in range(n)]

if mang_dx(n, arr):
  print("TRUE")
else:
  print("FALSE")
```

```python
def mang_dx(n, arr):
  mid = (n + 1) // 2
  count = 0
  hash_map = {}

  for i in range(n):
    if arr[i] not in hash_map:
      hash_map[arr[i]] = 0
    hash_map[arr[i]] += 1

  for i in range(mid):
    if hash_map[arr[i]] != 2:
      count += 1

  if count <= 1:
    return True
  return False

n = int(input())
arr = [int(input()) for _ in range(n)]

if mang_dx(n, arr):
  print("FALSE")
else:
  print("TRUE")
```

# **Count right turns in a list of points**

```python
def count_right_turns(n, points):
    right_turns = 0
    for i in range(1, n - 1):
        x1, y1 = points[i - 1]
        x2, y2 = points[i]
        x3, y3 = points[i + 1]

        cross_product = (x2 - x1) * (y3 - y2) - (y2 - y1) * (x3 - x2)
        print(cross_product)
        if cross_product < 0:
            right_turns += 1
    return right_turns

n = int(input())
points = [tuple(map(int, input().split())) for _ in range(n)]

print(count_right_turns(n, points))
```
